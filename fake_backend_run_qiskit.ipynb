# Circuit - 1

```python
from qiskit import QuantumCircuit, transpile
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.visualization import plot_histogram

# Generate a 5-qubit simulated backend
backend = GenericBackendV2(num_qubits=5)

# Create a simple circuit
circuit = QuantumCircuit(3)
circuit.h(0)
circuit.cx(0,1)
circuit.cx(0,2)
circuit.measure_all()
circuit.draw('mpl')
```

```python
# Transpile the ideal circuit to a circuit that can be directly executed by the backend
transpiled_circuit = transpile(circuit, backend)
transpiled_circuit.draw('mpl')
```

```python
# Run the transpiled circuit using the simulated backend
job = backend.run(transpiled_circuit)
counts = job.result().get_counts()
plot_histogram(counts)
```

# Circuit - 2

```python
# Create a sample circuit
qc = QuantumCircuit(4)
qc.h(0)
qc.cx(0, 1)
qc.cx(2, 3)
qc.measure_all()
qc.draw(output='mpl')
```

```python
# Transpile the ideal circuit to a circuit that can be directly executed by the backend
transpiled_circuit = transpile(qc, backend)
transpiled_circuit.draw('mpl')
```

```python
job = backend.run(transpiled_circuit)
counts = job.result().get_counts()
plot_histogram(counts)
```

# Circuit - 3

```python
qc_ip = QuantumCircuit(3,1) # 3 qubits, 1 classical bit

qc_ip.h(0)
qc_ip.cswap(0, 1, 2)
qc_ip.barrier() # Add a barrier for better visualization
qc_ip.measure(0,0) # Measure qubit 0 into classical bit 0 qc.measure(qubit_index, classical_bit_index)
qc_ip.draw('mpl')
```

```python
# Transpile the ideal circuit to a circuit that can be directly executed by the backend
transpiled_circuit = transpile(qc_ip, backend)
transpiled_circuit.draw('mpl')
```

```python
job = backend.run(transpiled_circuit)
counts = job.result().get_counts()
plot_histogram(counts)
```
